```
from pyspark.sql.functions import col, upper, substring, row_number, lpad
from pyspark.sql.window import Window

def agregar_codigo(df, prefijo_inicio=1):
    # Ordenar alfabéticamente para asignar número secuencial
    window = Window.orderBy("Municipios")

    # Crear iniciales y número secuencial
    df_codificado = df.withColumn("iniciales", upper(substring("Municipios", 1, 2))) \
        .withColumn("numero", row_number().over(window) + prefijo_inicio - 1) \
        .withColumn("codigo_municipio", 
                    col("iniciales") + 
                    lpad(col("numero").cast("string"), 2, "0")) \
        .drop("iniciales", "numero")

    return df_codificado

```

###  Consulta: Comparar la cantidad de deserción escolar en los municipios que aparecen tanto en Antioquia como en Santander

```
%sql
SELECT 
    a.Municipios,
    a.Cantidad AS Desercion_Antioquia,
    c.Cantidad AS Desercion_Cundinamarca,
    (a.Cantidad - c.Cantidad) AS Diferencia_Desercion
FROM Municipios_Antioquia a
JOIN Municipios_Cundinamarca c ON a.Municipios = c.Municipios
ORDER BY Diferencia_Desercion DESC;

```

### comparacion entre Primaria y Secundaria

```
%sql
SELECT 
    a.Municipios,
    a.Primaria AS Primaria_Antioquia,
    c.Primaria AS Primaria_Cundinamarca,
    a.Secundaria AS Secundaria_Antioquia,
    c.Secundaria AS Secundaria_Cundinamarca
FROM Municipios_Antioquia a
JOIN Municipios_Cundinamarca c ON a.Municipios = c.Municipios;
```

### los municipos que tienen una desercion mayo que cien

```
%sql
SELECT *
FROM (
    SELECT 
        a.Municipios,
        a.Cantidad AS Desercion_Antioquia,
        c.Cantidad AS Desercion_Cundinamarca,
        (a.Cantidad - c.Cantidad) AS Diferencia_Desercion
    FROM Municipios_Antioquia a
    JOIN Municipios_Cundinamarca c ON a.Municipios = c.Municipios
) t
WHERE ABS(Diferencia_Desercion) > 100;
```

### los municipios que tiene una desercion menor a 100

```
%sql
SELECT *
FROM (
    SELECT 
        a.Municipios,
        a.Cantidad AS Desercion_Antioquia,
        c.Cantidad AS Desercion_Cundinamarca,
        (a.Cantidad - c.Cantidad) AS Diferencia_Desercion
    FROM Municipios_Antioquia a
    JOIN Municipios_Cundinamarca c ON a.Municipios = c.Municipios
) t
WHERE ABS(Diferencia_Desercion) < 100;
```
